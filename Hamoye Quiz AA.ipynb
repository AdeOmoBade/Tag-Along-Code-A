```python
import pandas as pd
df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding='latin-1')
df.head(10)
```

```python
df.tail(10)
```

```python
wine_produced = df[df['Item'] == 'Wine'].groupby('Item')[['Y2015', 'Y2018']].sum()
print("Sum of Wine produced in 2015 is", wine_produced['Y2015'].values[0])
print("Sum of Wine produced in 2018 is", wine_produced['Y2018'].values[0])
```

```python
corr_element_code = {}
for year in df.columns[df.columns.str.startswith('Y')]:
    corr = df['Element Code'].corr(df[year])
    corr_element_code[year] = corr

min_corr_year = min(corr_element_code, key=corr_element_code.get)
min_corr_value = corr_element_code[min_corr_year]

print("Least correlation year with 'Element Code' is ", min_corr_year)
```

```python
total_missing_14 = df['Y2014'].isnull().sum()
total_entries_14 = df['Y2014'].shape[0]
percent_missing_14 = ((total_missing_14 / total_entries_14) * 100)
print(f"Total number of missing data in 2014 is {total_missing_14} and percentage is {'{:.3f}%'.format(percent_missing_14)}")
```

```python
stock_var = df[df['Element'] == 'Stock Variation'].groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum()
max_sum_year = stock_var.idxmax(axis=1).iloc[0]
print("The year with the highest sum of Stock Variation is", max_sum_year)
```

```python
protein_supply_15 = df[(df['Area'] == 'Madagascar') & (df['Element'] == 'Protein supply quantity (g/capita/day)')]['Y2015'].sum()
print("Protein supply quantity in Madagascar in 2015 is", round(protein_supply_15,2))
```

```python
y17 = df['Y2017']
mean_ = y17.mean()
std_ = y17.std()
print("Mean for 2017 is", round(mean_, 2))
print("Standard Deviation for 2017 is", round(std_, 2))
```
